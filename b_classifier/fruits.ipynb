```python
weight_df = f_df.loc[:, 'weight'].reset_index()
weight_df
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
std_weight = std.fit_transform(weight_df[['weight']])
std_weight
```

```python
weight_df.weight = std_weight
weight_df
```

```python
condition1 = weight_df.weight >= -1.96
condition2 = weight_df.weight <= 1.96
condition = condition1 & condition2

f_df = f_df.iloc[weight_df[condition].index]
f_df
```

```python
f_enc_df = f_df.reset_index()
f_enc_df
```
